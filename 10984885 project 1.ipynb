```python
mean_FA=data.fixedacidity.mean()
mean_CA=data.fixedacidity.mean()
mean_CL=data.fixedacidity.mean()
mean_FSO=data.fixedacidity.mean()
mean_TSO=data.fixedacidity.mean()
mean_D=data.fixedacidity.mean()
mean_pH=data.fixedacidity.mean()
mean_quality=data.fixedacidity.mean()
```

```python
print(mean_FA)
print(mean_CA)
print(mean_CL)
print(mean_FSO)
print(mean_TSO)
print(mean_D)
print(mean_pH)
print(mean_quality)
```

The above code was written to know the mean of column described in the variable and also it will help us to fill the null data columns with their mean.

```python
data['fixedacidity'] = data['fixedacidity'].fillna(mean_FA)
data['citricacid'] = data['citricacid'].fillna(mean_CA)
data['chlorides'] = data['chlorides'].fillna(mean_CL)
data['free_sulfurdioxide'] = data['free_sulfurdioxide'].fillna(mean_FSO)
data['total_sulfurdioxide'] = data['total_sulfurdioxide'].fillna(mean_TSO)
data['density'] = data['density'].fillna(mean_D)
data['pH'] = data['pH'].fillna(mean_pH)
data['quality'] = data['quality'].fillna(mean_quality)
```

```python
data
```

```python
mean=np.array([8.319875,0.271175,15.886875,46.52,0.9967456437500002,3.31110625,5.63625])
myLabels=['fixedacidity','citricacid','freeSO','totalSO','density','pH','quality',]
explode=(0,0.3,0,0,0.3,0,0)

plt.figure(figsize=(7,10))
plt.title('Mean Of specific columns', fontsize=9, fontweight='bold')
plt.style.use('ggplot')
plt.pie(mean,labels=myLabels,explode=explode,startangle=0,autopct='%.2f%%')
plt.legend(loc='upper left', fontsize=7)
plt.show()

```

### __DEDUCTION FROM PIE CHART__ 

The pie chart shows the average value of pH,density,quality,total sulfur dioxide, free sulfur dioxide,fixed acidity and citric acid of the beverage produced.

With the higher total sulfurdioxide showing in the chart it implies that, there is a high concentration of sulfur dioxide.
This means, this beverage is not good for asthmatics as the higher concentration of sulfur serves as an antioxidant.

**pH____** the pH shows the acidic content of the beverage and since it has an average value less than 7 it means that,the beverage has an acidic content which is high(**3.31**).
